## Script limpieza test file

Se muestra el proceso y ajuste de archivo de test, preprocesado de datos acorde a los parámetros y requerimientos para que sea compatible con el modelo de solución.

```python
import pandas as pd

TEST_PATH = "/content/test.csv"

print("Cargando TEST desde:", TEST_PATH)
df_test = pd.read_csv(TEST_PATH)
print("Shape TEST:", df_test.shape)
df_test.head()
```

```python
!pip install unidecode

import pandas as pd
import numpy as np
from unidecode import unidecode

def clean_colname(col):
    col = unidecode(str(col))
    col = col.upper().strip()
    col = col.replace(" ", "_")
    col = col.replace("-", "_")
    col = col.replace("__", "_")
    return col

def clean_cell_value(val):
    if pd.isna(val):
        return val

    val = unidecode(str(val))
    val = val.upper().strip()
    val = val.replace("-", "_")
    val = val.replace("  ", " ")
    return val

df_test.columns = [clean_colname(c) for c in df_test.columns]

text_cols = df_test.select_dtypes(include=["object"]).columns.tolist()

for c in text_cols:
    df_test[c] = df_test[c].apply(clean_cell_value)
```

```python
df_test = df_test.fillna(df_test.median(numeric_only=True))
```

```python
def show_unique_values(df, max_values=20):

    for col in df.columns:
        uniques = df[col].dropna().unique()
        total = len(uniques)

        print(f"Unique: {total}")

        if total <= max_values:
            print("Total:", uniques)
        else:
            print("Values:", uniques[:max_values], "...")
```

```python
show_unique_values(df_test)
df_test.columns.tolist()
```

```python
cat_cols = df_test.select_dtypes(include=["object", "category"]).columns.tolist()

for col in cat_cols:
    moda = df_test[col].mode()[0]
    df_test[col] = df_test[col].fillna(moda)
```

```python
# Verificar NaN
nan_counts = df_test.isna().sum()
nan_counts = nan_counts[nan_counts > 0]
print(nan_counts)
```

```python
OUTPUT_PATH = "/content/test_clean_adj.csv"

df_test.to_csv(OUTPUT_PATH, index=False, encoding="utf-8")
```
